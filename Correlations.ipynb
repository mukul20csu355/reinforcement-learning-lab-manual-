```python
import numpy as np
```

```python
np.random.seed(1)
```

```python
# 1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)
```

```python
x
```

```python
y = x + np.random.normal(0, 10, 1000)
y
```

```python
np.corrcoef(x, y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')
plt.scatter(x, y)
plt.show()
```

```python
y = 100 - x + np.random.normal(0, 5, 1000)
np.corrcoef(x,y)
```

```python
plt.scatter(x, y)
plt.xlabel('Values of X --->')
plt.ylabel('Values of Y --->')
plt.show()
```

```python
x = np.random.randint(0, 50, 1000)
y = np.random.randint(0, 50, 1000)
np.corrcoef(x, y)
```

```python
plt.scatter(x,y)
plt.xlabel('Values of X --- >')
plt.ylabel('Values of Y --- >')
plt.show()
```

```python
import pandas as pd 

df = pd.DataFrame({'a' : np.random.randint(0,50,1000)})
df['b'] = df['a'] + np.random.normal(0, 10, 1000) # positively correlated with 'a'
df['c'] = 100 - df['a'] + np.random.normal(0,5,1000) # negatively correlated with 'a'
df['d'] = np.random.randint(0, 50, 1000) # not correlated with 'a'

df.corr()
```

```python
pd.plotting.scatter_matrix(df, figsize = (16,16))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)), df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show()
```
